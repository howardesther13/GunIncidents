#### 4. Data Wrangling

```python
pd.set_option('display.max_columns', None)
```

```python
# separate date column into day_of_week, day, month and year
def separate_date_column(df, date_column):
    df[date_column] = pd.to_datetime(df[date_column])
    
    df['day_of_week'] = df[date_column].dt.day_name()
    df['day'] = df[date_column].dt.day
    df['month'] = df[date_column].dt.month
    df['year'] = df[date_column].dt.year
```

```python
separate_date_column(df, 'date')
```

```python
#Drop Unnamed: 29 column
df = df.drop(columns=['Unnamed: 29'])
df.head()

## -> drop all columns together
```

```python
#Drop participant_name and notes columns due to PII 
df = df.drop(columns=['notes', 'participant_name'])
df.head()
```

```python
#Drop sources, incident_url and source_url columns since they are not needed for analysis
df = df.drop(columns=['incident_url', 'source_url', 'sources'])
df.head()
```

```python
df['incident_url_fields_missing'].value_counts()
```

```python
# Drop incident_url_fields_missing since it does not contain any information
df = df.drop(columns=['incident_url_fields_missing'])
df.head()
```

```python
# separate participant_age_group column into adult, teen and child columns
def separate_age_group_column(df, age_group_column):
    df['n_adult'] = df[age_group_column].str.count('Adult')
    df['n_teen'] = df[age_group_column].str.count('Teen')
    df['n_child'] = df[age_group_column].str.count('Child')
    return df
```

```python
# separate gender column into Female and Male columns
def separate_gender_column(df, gender_column):
    df['n_male'] = df[gender_column].str.count('Male')
    df['n_female'] = df[gender_column].str.count('Female')
    return df
```

```python
# separate type column into Victim and Suspect 
def separate_type_column(df, type_column):
    df['n_victim'] = df[type_column].str.count('Victim')
    df['n_suspect'] = df[type_column].str.count('Suspect')
    return df
```

```python
# create n_participants column from participant_status column
def create_n_participants(df, status_column):
    df['n_participants'] = df[status_column].str.count('\:\:')
    return df
```

```python
# Apply functions to dataframe
df = separate_age_group_column(df, 'participant_age_group')
df = separate_gender_column(df, 'participant_gender')
df = separate_type_column(df, 'participant_type')
df = create_n_participants(df, 'participant_status')
df.head()
```

```python
# Drop unnecessary columns after functions
df = df.drop(columns=['gun_stolen', 'gun_type', 'incident_characteristics', 'participant_age', 'participant_age_group',
                      'participant_gender', 'participant_relationship', 'participant_status', 'participant_type'])
df.head()
```

```python
# Reorganize columns
df = df[['incident_id', 'date', 'day_of_week', 'day', 'month', 'year', 'address', 'city_or_county', 'state', 'n_participants',
        'n_guns_involved', 'n_killed', 'n_injured', 'n_adult', 'n_teen', 'n_child', 'n_male', 'n_female', 'n_victim', 
         'n_suspect','location_description', 'latitude', 'longitude', 'state_house_district', 'state_senate_district', 
         'congressional_district']]
df.head()
```
